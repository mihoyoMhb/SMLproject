```python
import pandas as pd
data = pd.read_csv('data/training_data_fall2024.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(x='hour_of_day', hue='increase_stock', data=data, palette='viridis')
plt.xlabel('Hour of Day')
plt.ylabel('Count')
plt.title('Bike Demand by Hour of Day')
plt.legend(title='Bike Demand', labels=['Low', 'High'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='day_of_week', hue='increase_stock', data=data, palette='viridis')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.title('Bike Demand by Day of Week')
plt.legend(title='Bike Demand', labels=['Low', 'High'])
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(x='month', hue='increase_stock', data=data, palette='viridis')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Bike Demand by Month')
plt.legend(title='Bike Demand', labels=['Low', 'High'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='holiday', hue='increase_stock', data=data, palette='viridis')
plt.xlabel('Holiday (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.title('Bike Demand on Holidays vs. Non-Holidays')
plt.legend(title='Bike Demand', labels=['Low', 'High'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='increase_stock', y='snowdepth', data=data, palette='viridis')
plt.xlabel('Bike Demand (0 = Low, 1 = High)')
plt.ylabel('Snow Depth (cm)')
plt.title('Snow Depth Distribution by Bike Demand')
plt.show()
```

```python
numerical_features = ['temp', 'dew', 'humidity', 'precip', 'snowdepth', 'windspeed', 'cloudcover', 'visibility']

for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[feature], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```
